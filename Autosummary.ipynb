```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
articleURL = "https://www.theverge.com/2017/2/3/14495606/beoplay-h9-review-wireless-noise-canceling-headphones"
```

```python
page = urllib.request.urlopen(articleURL).read().decode('utf8', 'ignore')
soup = BeautifulSoup(page, "lxml")
soup
```

```python
soup.find_all('p')
```

```python
text = ' '.join(map(lambda p: p.text, soup.find_all('p')))
text
```

```python
text.encode('ascii', errors='replace')
text.replace("?", " ")
```

```python
def getTextFromUrl(url, tag):
    page = urllib.request.urlopen(articleURL).read().decode('utf8', 'ignore')
    soup = BeautifulSoup(page, "lxml")
    text = ' '.join(map(lambda p: p.text, soup.find_all(tag)))
    text.encode('ascii', errors='replace')
    return text.replace("?", " ")
```

```python
text = getTextFromUrl(articleURL, "p")
print(text)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sents = sent_tokenize(text)
sents
```

```python
word_sent = word_tokenize(text.lower())
word_sent
```

```python
_stopwords = set(stopwords.words('english') + list(punctuation))
_stopwords.add("â€™")
```

```python
word_sent=[word for word in word_sent if word not in _stopwords]
word_sent
```

```python
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
```

```python
from heapq import nlargest
```

```python
nlargest(10, freq, key=freq.get)
```

```python
from collections import defaultdict
ranking = defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        ranking[i] += freq[w]
        
ranking
```

```python
sents_xid = nlargest(4, ranking, key=ranking.get)
sents_xid
```

```python
[sents[j] for j in sorted(sents_xid)]
```
